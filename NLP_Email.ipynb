```python
df["type"] = df["type"].apply(lambda x: 1 if x == "spam" else 0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['text'],df['type'],test_size=0.3)
```

```python
print(x_train)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

v=TfidfVectorizer()
x_train_count=v.fit_transform(x_train.values)
```

```python
x_train_count.toarray()[:2]
```

model = MultinomialNB()
model.fit(x_train_count, y_train)

```python
emails = ["hello, how are you", "Big Discount sales startsat 10% discount"]
email_count = v.transform(emails)
model.predict(email_count)
```
